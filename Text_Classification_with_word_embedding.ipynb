```python
import pandas as pd
import numpy as np
```

```python
SMS_SPAM = pd.read_csv("dataset_sms_spam_v1.csv")
```

```python
from nltk.stem import PorterStemmer
words = SMS_SPAM['Teks'][0].split()
ps = PorterStemmer()
for w in words:
    roodWord = ps.stem(w)    
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
tokenization = nltk.word_tokenize(roodWord)
for w in tokenization:
    print (tokenization)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_tf = count_vect.fit_transform(SMS_SPAM['Teks'])
print("Shape of term-frequency matrix:", X_train_tf.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_tf)
print("Shape of term frequency-inverse document frequency matrix:",X_train_tfidf.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

mod1 = MultinomialNB()
mod1.fit(X_train_tf, SMS_SPAM['Teks'])
print(mod1)
```

```python
mod2 = MultinomialNB()
mod2.fit(X_train_tfidf, SMS_SPAM['Teks'])
print(mod2)
```

```python
X_test_tf = count_vect.transform(SMS_SPAM['Teks'])
X_test_tfidf = tfidf_transformer.transform(X_test_tf)
```

```python
predicted1 = mod1.predict(X_test_tf)
predicted2 = mod2.predict(X_test_tfidf)
```

```python
print ("Accuracy:", accuracy_score(SMS_SPAM['Teks'],predicted1))
print ("Accuracy:", accuracy_score(SMS_SPAM['Teks'],predicted2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(SMS_SPAM['Teks'],predicted1,target_names=SMS_SPAM['Teks']))
print(classification_report(SMS_SPAM['Teks'],predicted2,target_names=SMS_SPAM['Teks']))
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize 
data = []

# iterate through each sentence in the file 
for i in SMS_SPAM['Teks']:
    temp = []
    
    # tokenize the sentence into words
    for j in word_tokenize(i):
        temp.append(j.lower())
        
    data.append(temp)
```

```python
import gensim 
model = gensim.models.Word2Vec(data, min_count = 1, size = 50, window = 3)
```

```python
model.save('word2vec_model_file.model')
```

```python
def get_mean_vector(word2vec_model, words):
    # remove out-of-vocabulary words
    words = [word for word in words if word in word2vec_model.wv.vocab]
    if len(words) >= 1:
        return np.mean(word2vec_model[words], axis=0)
    else:
        return []
```

```python
vec = get_mean_vector(model,data[0])
vec
```

```python
mat = []
for doc in data:
    vec = get_mean_vector(model,doc)
    mat.append([vec,doc])
```
